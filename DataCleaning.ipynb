##### First we will take a look at our stock data

```python
df = pd.read_csv('Data/stock_info_complete.csv')
df.info()
```

```python
df.drop('Unnamed: 0', axis=1,inplace=True)
df.date = pd.to_datetime(df.date)
```

```python
df.describe()
```

For some reason it looks like we have negative days post ipo. We will recalculate these values by subtracting date entries from each stocks first listed date

```python
ticker_dates = df.groupby('ticker')['date'].min()
df['days_post_ipo'] = df.apply(lambda row: (row.date - ticker_dates[row.ticker]).days, axis=1)
df.describe()
```

#### Interesting we have adjusted closing prices that are negative
- How many rows does this impact?
- Why could this be?

```python
print(len(df[df.adjclose <0]))
```

[Here](https://www.investopedia.com/terms/a/adjusted_closing_price.asp#:~:text=The%20adjusted%20closing%20price%20amends,detailed%20analysis%20of%20past%20performance), we can see that adjusted close price takes into account dividends and splits.
<br>

This means that the values are correct but this variable will probably not be suitable for our goals here.
<br>

We will drop these features to avoid confusion later.

```python
df.drop('first_day_adjclose', axis = 1, inplace=True)
df.drop('adjclose', axis = 1, inplace=True)
df.describe()
```

##### Let's update the csv for later use

```python
df.to_csv('Data/stock_info_cleaned.csv')
```

### This seems more manageable to start converting features

```python
df.info()
```

- It looks like most of our data is formatted as strings
- Also, we have 2 variables with few non null values (we will drop those)

```python
df.drop('filedDate', axis=1, inplace=True)
df.drop('expectedPriceDate', axis=1, inplace=True)
```

```python
df.Employees = df.Employees.str.split(' ', expand=True)[0]
df = df.replace('--', '')
quant_cols = ['Offer amount','Total Offering Expense',"Stockholders' Equity",'Shareholder Shares Offered','Shares Outstanding', 'Revenue','Net Income', 'Total Assets',
       'Total Liabilities','proposedSharePrice', 'sharesOffered', 'dollarValueOfSharesOffered']
df[quant_cols] = df[quant_cols].replace('[^.0-9\(-)]', '', regex=True
       ).replace('\(', '', regex=True).replace('\)', '', regex=True
       ).replace('', np.nan, regex=True)
df.replace('NaN', np.NaN)
df.replace('', np.NaN)
df.info()
```

### Let's convert try converting columns to floats and see which ones give us errors

```python
for col in df.columns:
    try:
        df[col] = df[col].astype(float)
    except:
        print(f'Could not convert column:\t{col}')
df.info()
```

- Looks like most of the transoformations we wanted came through. Let's investigate the ones that did not

```python
# df[df['proposedSharePrice'].apply(type) ==str]
# df['proposedSharePrice'].apply(type)
# df['proposedSharePrice'] = df['proposedSharePrice'].astype(str)
indices = df[df['proposedSharePrice'].str.count('\.') >1].index
df['proposedSharePrice'].iloc[indices] = np.NaN
```

```python
df['Shares Over Alloted']=  df['Shares Over Alloted']!= ''
df['Shares Over Alloted']=  df['Shares Over Alloted']*1
df.head()
```

```python
for col in df.columns:
    try:
        df[col] = df[col].astype(float)
    except:
        print(f'Could not convert column:\t{col}')
df.info()
```

- Drop a few more columns

```python
df.drop('Fiscal Year End', axis=1, inplace=True)
df.drop('Company Phone', axis=1, inplace=True)
df.drop('proposedExchange', axis=1, inplace=True)
df.drop('companyName', axis=1, inplace=True)
df.head()
```
