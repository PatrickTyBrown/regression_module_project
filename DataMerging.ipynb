```python
import pickle
import pandas as pd
```

```python
def pickle_file(obj, path):
    file = open(path, 'wb')
    pickle.dump(obj, file)
    file.close()

def load_pickle(path):
    file = open(path, 'rb')
    obj = pickle.load(file)
    file.close()
    return obj
```

```python
def offset_column(df, column, days):
    df.sort_values(by=['ticker', 'date'], inplace=True)
    list_shifted = []
    for ticker in df.ticker.unique():

        list_shifted.append( df[(df.ticker == ticker)][column].shift(days))

    column_name = f'{column}_{days}_out'
    df[column_name] = pd.concat(list_shifted)
    return df

df = offset_column(df, 'close', 5)
df.head()
```
