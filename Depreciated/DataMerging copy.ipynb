```python
import pickle
import pandas as pd
```

```python
def pickle_file(obj, path):
    file = open(path, 'wb')
    pickle.dump(obj, file)
    file.close()

def load_pickle(path):
    file = open(path, 'rb')
    obj = pickle.load(file)
    file.close()
    return obj
```

```python
ipo_open_keys= {}
unique_tickers = list(stock_info.ticker.unique())
for i in range(len(unique_tickers)):
    ipo_open_keys[unique_tickers[i]]  =  stock_info.loc[stock_info.ticker == unique_tickers[i]].iloc[0][['open', 'high', 'low', 'close',
       'adjclose', 'volume']]

# stock_info['price_change'] = stock_info.apply(lambda row: None if row.ticker not in unique_tickers else
#                     row.adjclose / ipo_open_keys[row.ticker], axis=1)
# stock_info.to_csv('Data/stock_info')
# stock_info.tail(10)
```

```python
columns = ['open', 'high', 'low', 'close',
       'adjclose', 'volume']

for col in columns:
    stock_info['first_day_' + col] = stock_info.apply(lambda row: None if row.ticker not in unique_tickers else
                        row[col] / ipo_open_keys[row.ticker][col], axis=1)
stock_info.to_csv('Data/stock_info_complete')
stock_info.tail(10)
```

```python
# stock_info.loc[stock_info.ticker == 'MEAD'].iloc[0][['open', 'high', 'low', 'close',
#        'adjclose', 'volume']]['open']

first_day_df = pd.DataFrame(ipo_open_keys).transpose().reset_index()
first_day_df.columns = ['ticker', 'first_day_open', 'first_day_high', 'first_day_low', 'first_day_close',
       'first_day_adjclose', 'first_day_volume']
first_day_df.head()
```

```python
stock_data = stock_info.merge(first_day_df, how='left', on='ticker')
stock_data.head()
```

```python
stock_data.drop('price_change', axis=1, inplace=True)
stock_data.drop('Unnamed: 0.1', axis=1, inplace=True)
stock_data.drop('Unnamed: 0', axis=1, inplace=True)
stock_data.head()
```
